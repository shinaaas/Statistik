# Expression and Variables

**Expression** dalam Python dapat mencakup operasi di antara jenis yang kompatibel (mis., Bilangan bulat dan pecahan). Misalnya, operasi aritmatika dasar seperti menambahkan beberapa angka, misalnya:

```python
# Addition operation expression

43 + 60 + 16 + 41
```

```python
# Subtraction operation expression

50 - 60
```

```python
# Multiplication operation expression

5 * 5
```

```python
# Division operation expression

25 / 7
```

```python
# Integer division operation expression (Modulus)

25 // 7
```

Sama seperti kebanyakan bahasa pemrograman, kita dapat menyimpan nilai dalam **variabel**, sehingga kita bisa menggunakannya nanti dalam program.

```python
# Store value into variable

x = 43 + 60 + 16 + 41
```

Untuk melihat nilai x dalam program, kita cukup menuliskan nama variabel tersebut dan meletakkannya di baris terakhir sel

```python
# Print out the value in variable

x
```

Menggunakan variabel untuk menyimpan nilai hasil dari operasi.

```python
# Use another variable to store the result of the operation between variable and value

y = x / 60
y
```

Kita juga dapat menimpa nilai dari suatu variabel dengan nilai yang baru yang didapat dari sebuah operasi ataupun yang lainnya.

```python
# Overwrite variable with new value

x = x / 60
x
```

Untuk praktek ada baiknya untuk menggunakan nama variabel yang bermakna atau memiliki arti, sehingga kita dan orang lain dapat membaca kode dan memahaminya dengan lebih mudah, contoh:

```python
# Name the variables meaningfully

total_min = 43 + 42 + 57 # Total length of albums in minutes
total_min
```

```python
# Name the variables meaningfully

total_hours = total_min / 60 # Total length of albums in hours 
total_hours
```

## <h3 id="operations">Basic String Operations</h2>

Ada banyak metode operasi string dalam Python yang dapat digunakan untuk memanipulasi data. Berikut beberapa operasi string dasar pada data.

Mari kita coba dengan method upper; metode ini mengonversi karakter huruf kecil ke karakter huruf besar.

```python
# Convert all the characters in string to upper case

A = "Thriller is the sixth Michael Jackson studio album"
print("before upper:", A)
B = A.upper()
print("After upper:", B)
```

Method replace menggantikan segmen string, yaitu substring dengan string baru. Kita memasukkan bagian dari string yang ingin kita ubah. Argumen kedua adalah dengan apa kami ingin bertukar segmen, dan hasilnya adalah string baru dengan segmen yang diubah:

```python
# Replace the old substring with the new target substring is the segment has been found in the string

C = A.replace('Michael', 'Janet')
C
```

Method find() digunakan untuk menemukan sebuah string dari string yang ada dan mengembalikan nilai index dimana string tersebut ditemukan.

```python
# Find the substring in the string. Only the index of the first elment of substring in string will be the output

Name = "Michael Jackson"
Name.find('el')
```

```python
# Find the substring in the string.

Name.find('Jack')
```
