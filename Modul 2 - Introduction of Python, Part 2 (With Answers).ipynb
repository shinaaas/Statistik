## Tuple, List, & Set



Dalam Python, ada berbagai tipe data: string, integer dan float. Tipe data ini semua dapat disimpan didalam suatu tempat bernama **Tuple**, **List** dan juga **Set**. Lalu apakah perbedaan dari ketiganya?

**Tuple** adalah sebuah tempat penyimpanan data yang dimana memilki sifat **Immutable** yang berarti nilai yang terdapat di dalam **Tuple** tidak dapat dirubah lagi setelah di deklarasikan. Penulisan **Tuple** menggunakan tanda buka dan tutup kurung biasa.

```python
# Membuat Tuple

tuple1 = ("disco lazy time", 10, 1.2 )
tuple1
```

**Tuple** bersifat ordered, yang dimana berarti **Tuple** dapat dipanggil menggunakan konsep indexing dan untuk urutannya dapat dilihat dari data pertama yang masuk.

```python
# Print setiap index dari tuple

print(tuple1[0])
print(tuple1[1])
print(tuple1[2])
```

Percobaan merubah isi **Tuple**.

```python
# Percobaan merubah isi Tuple

tuple1[1] = "pop"
```

**List** sama dengan Tuple adalah sebuah simple collection object yang dapat menampung berbagai macam tipe data ataupun object di dalamnya. Perbedaannya dengan Tuple adalah **List** bersifat Mutable yang artinya isi dari **List** dapat dirubah kapan pun.

Untuk penulisan dari **List** menggunakan buka dan tutup kurung siku.

```python
# Membuat sebuah list

List = ["Michael Jackson", 10.1, 1982]
List
```

**List** dapat berisi string, float, dan integer. Kita juga dapat membuat nested list, membuat tuple dan struktur data lainnya pada **List**.

```python
# Membuat sebuah list dengan isi nested list dan Tupple

List = ["Michael Jackson", 10.1, 1982, [1, 2], ("A", 1)]
List
```

Untuk menambahkan data ke dalam list kita dapat menggunakan 2 buah fungsi, yaitu **append()** dan **extend()**.

Perbedaan antara **append()** dan **extend()** terletak pada jumlah data yang ditambahkannya. Untuk lebih jelasnya dapat dilihat pada baris kode dibawah:

```python
# Menggunakan extend untuk menambah data

List = [ "Michael Jackson", 10.2]
List.extend(['pop', 10])
List
```

```python
# Menggunakan append untuk menambah data

List = [ "Michael Jackson", 10.2]
List.append(['pop', 10])
List
```

Dari kode diatas dapat dilihat perbedaan ketika menggunakan **append()** dan **extend()**, ketika menambahkan 2 data melalui fungsi **extend()** maka data akan bertambah 2 sejumlah yang diinputkan, tetapi ketika menggunakan fungsi **append()** data yang ditambahkan akan digabung menjadi sebuah **list** lalu ditambahkan ke dalam **list** yang kita inginkan.

Karena **list** bersifat **Mutable**, maka kita dapat merubah isi data yang telah ada didalam **list**.

```python
print("List Before Change: ", List)
List[0] = "Janet Jackson"
print("List After Change: ", List)
```

Kita juga dapat menghapus anggota data dari **list** menggunakan fungsi **del()**.

```python
print('Before change:', List)
del(List[0])
print('After change:', List)
```

**Set** adalah sebuah unique collection object yang ada dalam bahasa pemrograman Python. Sama halnya dengan **Tuple** dan **List**, kita dapat menyimpan banyak data dengan tipe data berbeda dalam satu **Set**. Kita dapat merepresentasikan sebuah **Set** dengan tanda kurung kurawal.

**Set** bersifat unordered sehingga tidak memungkinkan untuk melakukan indexing jika menggunakan **Set**. **Set** juga bersifat **unique**, yang berarti setiap data didalam **Set** tidak boleh ada yang sama atau unik. Jika saat inisialisasi terdapat data yang sama maka Python otomatis akan menghapus data yang sama dan hanya menyisakan 1 data.

```python
# Create a set

set1 = {"pop", "EDM", "ballad", "ballad", "rock", "soul", "hard rock", "rock", "R&B", "rock", "disco"}
set1
```

Kita juga dapat membuat sebuah **Set** dengan mengkonversi **list** menjadi **Set** seperti contoh kode dibawah: 
(so you know the unique value)


```python
# Merubah list menjadi set

genres = ["pop", "pop", "rock", "folk rock", "hard rock", "soul",
                    "progressive rock", "soft rock", "R&B", "disco"]
music_genres = set(genres)
music_genres
```

```python
# Atau dengan cara

music_genres = set(["pop", "pop", "rock", "folk rock", "hard rock", "soul",
                    "progressive rock", "soft rock", "R&B", "disco"])
music_genres
```

### Set Logic Operation

Pada bagian ini kita akan mencoba melakukan fungsi logika pada **Set** yaitu **intersection**, **difference**, dan **union**.

```python
# Membuat sample set

twice_song1 = set(["Cheer Up", 'Fancy', 'TT', "Yes or Yes", "Signal"])
twice_song2 = set([ "Like Oh-Ah", "TT", "Likey", "What Is Love?", "Heartshaker", "Signal", "Cheer Up"])
```

Dari baris kode diatas kita telah membuat 2 buah set yang berisikan judul lagu dengan data masing masing set adalah sebagai berikut:

```python
twice_song1
```

```python
twice_song2
```

Dari 2 buah set yang telah dibuat kita akan mengambil data yang sama terdapat di 2 buah set tersebut dengan menggunakan fungsi **intersection**.

```python
intersection = twice_song1 & twice_song2
intersection
```

```python
twice_song1.intersection(twice_song2)
```

Dari kode diatas, telah didapatkan data yang termasuk ke dalam 2 set tersebut, yaitu: **Cheer Up**, **Signal** dan **TT**.

Selanjutnya kita dapat menampilkan data pada **twice_song1** yang tidak sama dengan data pada **twice song2**, begitupun sebaliknya menggunakan fungsi **difference()**.

```python
twice_song1.difference(twice_song2)
```

```python
twice_song2.difference(twice_song1)
```

Selanjutnya, untuk memeriksa union atau gabungan dari 2 set dapat menggunakan fungsi **union()**.

```python
twice_song1.union(twice_song2)
```

Terakhir, untuk memperiksa apakah sebuah data termasuk sub set dari suatu set dapat menggunakan fungsi **issubset()** dan jika ingin memperiksa apakah suatu elemen merupakan sebuah super set dari sebuah elemen dapat menggunakan fungsi **issuperset()**.

```python
# Penggunaaan issuperset

set(twice_song1).issuperset(twice_song2)
```

```python
# Cara lain penggunaan issuperset

set(twice_song1).issuperset({"TT", "Yes or Yes"})
```

```python
# Penggunaan issubset

set({"Like Oh-Ah", "TT"}).issubset(twice_song1)
```

```python
# Cara lain penggunaan issubset

set(twice_song1).issubset(twice_song2)
```

### Branching

**Branching** atau **Percabangan** memungkinkan kita untuk menjalankan pernyataan yang berbeda untuk setiap input yang berbeda.

Sebagai contoh jika pernyataan itu **Benar** kita dapat memasuki suatu opsi percabangan dan program akan menjalankan beberapa tugas yang telah ditentukan, tetapi jika pernyataan itu **Salah** program akan mengabaikan tugas ataupun menjalankan opsi percabangan lain.


**Single if statement**

Kita dapat menggunakan condition statement yang dipelajari sebelumnya untuk  memeriksa persyaratan dalam statement if.

Sintaksnya adalah **if condition statement:**. Tulis kode yang perlu dijalankan dalam kondisi ini di baris baru dengan indent. Baris kode setelah titik dua dan dengan indent yang hanya akan dieksekusi ketika pernyataan if bernilai **True**. Kode akan berakhir ketika baris kode tidak mengandung indent.

```python
# Contoh if statement

age = 19

# Ekspresi yang dapat menghasilkan nilai True atau False
if age > 18:
    
    # Baris baru menggunakan indent, kita akan menjalankan baris kode jika kondisi bernilai True
    print("you can enter" )

# Statement dibawah ini akan berjalan tanpa memperdulikan hasil dari statement if
print("have fun")
```

Dalam kasus diatas ini, kode yang dijalankan adalah untuk mencetak ***("you can enter")*** hanya jika variabel age bernilai lebih dari 18 karena baris kode ini memiliki indent. Namun, kode untuk mencetak ***("have fun")*** akan selalu dieksekusi dan tidak akan terpengaruh oleh pernyataan if.

**If-else statement**

Statement else akan menjalankan blok kode jika tidak ada kondisi yang Benar sebelum statement else ini, dengan kata lain statement else berguna sebagai tempat menampung untuk kondisi bernilai **Salah**. 

Sintaks pernyataan else sama dengan sintaks pernyataan if, dengan penulisan **else:**.


```python
# Contoh statement else

age = 18

if age > 18:
    print("you can enter")
else:
    print("you can't enter")
    
print("have fun")
```

**If-else if-else statement**

Statement elif merupakan kependekan dari else if, memungkinkan kita untuk memeriksa kondisi tambahan jika kondisi pada statement sebelumnya salah. Jika kondisi untuk pernyataan elif Benar, ekspresi alternatif tersebut akan dijalankan dan sisa statement yang lain akan dilewati.

Penulisan syntax untuk statement else if adalah **elif condition statement:**.

```python
# Contoh statement elif

age = 18

if age > 18:
    print("you can enter" )
elif age == 18:
    print("you can enter too")
else:
    print("you can't enter" )
    
print("have fun")
```

### Logical Operators

Terkadang kita ingin memeriksa lebih dari satu kondisi sekaligus. Misalnya, kita mungkin ingin memeriksa apakah satu kondisi dan kondisi lainnya Benar untuk menjalankan beberapa baris kode. Operator logika memungkinkan kita untuk menggabungkan atau mengubah kondisi, berikut adalah beberapa operator logika yang digunakan

### For Loops

**For loop** memungkinkan kita untuk menjalankan blok kode beberapa kali sesuai jumlah yang telah ditentukan. Misalnya, kita akan menggunakan ini untuk mencetak sebuah output sebanyak 3 kali ataupun menerima input sebanyak 3 kali.

```python
# Contoh for loop

dates = [1982,1980,1973]
N = len(dates)

for i in range(N):
    print(dates[i])     
```

```python
# Contoh lain penggunaan for loop

for i in range(0, 8):
    print(i)
```

Khusus pada Python kita dapat langsung mengakses jumlah elemen dari list dengan memanggil nama listnya langsung pada syntax **for loops** kita.

```python
# Contoh penggunaan for loops dengan langsung memanggil list

for year in dates:  
    print(year)   
```

Ataupun kita dapat menggunakan 2 parameter penghitung pada sebuah syntax **for loops** seperti pada contoh dibawah:

```python
# Contoh Penggunaan 2 paramater penghitung pada for loops

squares=['red', 'yellow', 'green', 'purple', 'blue']

for i, square in enumerate(squares):
    print(i, square)
```

### While Loops

Seperti yang kita lihat, for loops digunakan untuk aliran pengulangan yang terkontrol. Namun, bagaimana jika kita tidak tahu kapan kita ingin menghentikan loop? Bagaimana jika kita ingin terus menjalankan blok kode sampai kondisi tertentu terpenuhi?

**While loops** ada sebagai alat untuk menjalankan eksekusi berulang kali berdasarkan kondisi. Blok kode akan terus dieksekusi sampai kondisi logika yang diberikan mengembalikan nilai boolean False.

```python
# Contoh penggunaan while Loop

dates = [1982, 1980, 1973, 2000]

i = 0
year = 0

while(year != 1973):
    year = dates[i]
    i = i + 1
    print(year)

print("It took ", i ,"repetitions to get out of loop.")
```

## Functions in Python

**Function** atau **fungsi** adalah sebuah blok kode yang dapat digunakan kembali untuk melakukan tugas yang ditentukan dalam fungsi tersebut. 

Menggunakan fungsi membuat kita dapat memecah tugas dan memungkinkan kita untuk menggunakan kembali kode kita di berbagai program.

Ada dua jenis fungsi:
* **Pre-defined functions**
* **User defined functions**

Kita dapat menentukan fungsi untuk menyediakan kebutuhan fungsionalitas yang diperlukan. Berikut adalah aturan sederhana untuk mendefinisikan fungsi dengan Python:

* Blok fungsi dimulai dengan def lalu diikuti oleh nama fungsi dan tanda kurung ().
* Opsional untuk menempatkan parameter input atau argumen yang di dalam tanda kurung ini.
* Kita juga dapat mendefinisikan parameter di dalam tanda kurung ini.
* Ada body di dalam setiap fungsi yang dimulai dengan titik dua (:) dan diberi indent.
* Kita juga dapat menempatkan dokumentasi di dalam body fungsi
* Return keluar dari suatu fungsi, secara opsional memberikan nilai kembali ketempat fungsi dipanggil

Gambarannya dapat dilihat pada ilustrasi dibawah:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%203/Images/FuncsDefinition.png" width="500" /> 

```python
# Contoh membuat sebuah fungsi sederhana

def add(a):
    b = a + 1
    print(a, "if you add one then the result is", b)
    return(b)
```

```python
# Memanggil fungsi add()

add(1)
```

**Setting default argument values in your custom functions**

Kita juga dapat menetapkan nilai default untuk argumen di fungsi. Misalnya, dalam fungsi isGoodRating(), bagaimana jika kita ingin membuat ambang batas untuk apa yang kita anggap sebagai rating yang baik? Mungkin secara default, kita seharusnya memiliki rating default 4:

```python
# Example for setting param with default value

def isGoodRating(rating=4): 
    if(rating < 7):
        print("this album sucks it's rating is",rating)
        
    else:
        print("this album is good its rating is",rating)
```

```python
isGoodRating()
isGoodRating(10)
```
